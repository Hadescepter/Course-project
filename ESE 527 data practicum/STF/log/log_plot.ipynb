```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./log.csv")


plt.figure(figsize=(8, 5))
plt.plot(df["epoch"], df["loss"], marker='o', linewidth=2, color='royalblue')


plt.title("Qwen LoRA Training Loss vs Epoch", fontsize=14, fontweight='bold')
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Loss", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()


plt.show()
```
