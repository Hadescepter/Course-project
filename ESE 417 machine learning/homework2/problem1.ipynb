```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

Problem1-b

```python
# generate dataset
n=200
x=np.linspace(1,10,n)
noise=np.random.normal(0,1,n)
y=-5*x+15+noise
y_model=-5*x+15
```

```python
plt.plot(x,y_model,color='red',lw=2,label='model_line')
plt.scatter(x,y,color='blue',s=10,label='observed_sample')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
Sum_residuals_1=np.sum((y-y_model)**2)
print(Sum_residuals_1)
```

Prblem1-C

```python
#least-square
x_mean=x.mean()
y_mean=y.mean()
w1=(np.sum(x*y)-n*x_mean*y_mean)/(np.sum(x*x)-n*x_mean*x_mean)
w0=y_mean-w1*x_mean
y_least_square=x*w1+w0
```

```python
print(w1,w0)
```

```python
plt.plot(x,y_least_square,color='red',lw=2,label='least_squares_line')
plt.scatter(x,y,color='blue',s=10,label='observed_sample')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python

Sum_residuals=np.sum((y-y_least_square)**2)
print(Sum_residuals)
```

Problem1d

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
model = LinearRegression()
model.fit(x.reshape(-1,1), y.reshape(-1,1))
```

```python
w1_1=model.coef_[0][0]
W0_0=model.intercept_[0]
```

```python
print(w1_1,W0_0)
```

```python
y_sklearn=x*w1_1+W0_0
```

```python
plt.plot(x,y_sklearn,color='red',lw=2,label='SK_learn')
plt.scatter(x,y,color='blue',s=10,label='observed_sample')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
mse = mean_squared_error(y, y_sklearn)
rss=mse*n
print(rss)
```
